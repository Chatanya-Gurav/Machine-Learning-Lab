```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('Salary_Data.csv')
print(data1.head()) 
print(data1.describe())
```

```python
plt.title('Salary Distribution of Salary')

sns.regplot(x='YearsExperience',y='Salary',data=data1)
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

def calculate_rmse(actual, predicted):
    mse = mean_squared_error(actual, predicted)
    rmse = np.sqrt(mse)
    return rmse

data1 = pd.read_csv('Salary_Data.csv')
print(data1.head()) 
print(data1.describe())

X = data1[['YearsExperience']]
y = data1['Salary']

model = LinearRegression()
model.fit(X, y)

m = model.coef_[0]
c = model.intercept_
print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")

y_pred = model.predict(X)

mse = mean_squared_error(y, y_pred)
rmse = calculate_rmse(y, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")

plt.title('Salary Distribution of Salary')
sns.regplot(x='YearsExperience', y='Salary', data=data1, line_kws={"color": "red"})
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
